```python
import pandas as pd
import numpy as np
import os
%matplotlib inline
```

```python
# Levantamos los datos del disco
output = []

for fname in os.listdir('./knn_normal'):
    tag = fname.replace('statisticsknn_test_', '').replace('.out','')
    
    with open('./knn_normal/' + fname, 'r') as handle:
        data = handle.read().split('\n')
        
        for line in data:
            spl = line.split('\t\t\t')
            
            try:
                values = spl[1].split(' ')

                for index, value in enumerate(values):
                    if len(value) > 0:
                        value = int(value)
                        index = len(values) - index - 1

                        output.append({
                                'test': tag,
                                'metric': spl[0],
                                'index': index,
                                'value': value
                            })
            except:
                pass
```

```python
# Levantamos el data frame
df = pd.DataFrame(output)
```

```python
# Calculamos los intervalos de confianza
import scipy as sp

grp = df[df['metric'] == 'kNN Hit CV'].groupby('test')

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0*np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * sp.stats.t._ppf((1+confidence)/2., n-1)
    return m, m-h, m+h

cis = []
for (test, group) in grp:
    (mean_center, mean_lower, mean_upper) = mean_confidence_interval(group['value'].map(lambda x: float(x)/4200.0))
    
    cis.append({
            'test': int(test),
            'mean_lower': mean_lower,
            'mean_upper': mean_upper,
            'mean_center': mean_center
        })
    
cis.sort(lambda x, y: cmp(x['test'], y['test']))
```
